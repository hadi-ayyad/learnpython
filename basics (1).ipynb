# Functions

```python
def my_function():
    print("This is my function")
```

```python
my_function()
```

```python
def my_function_with_args(username, greeting):
    print("Hello, %s, From my_function()! I wish you %s" % (username, greeting))
```

```python
my_function_with_args("John", "a great year!")
my_function_with_args("Eric", "an awesome semester!")
my_function_with_args("Jessica", "the best of luck!")   
```

```python
names = ['John', 'Eric', 'Jessica']
greetings = ['a great year', 'an awesome semester', 'the best of luck'] 

for name, greeting in zip(names, greetings):
    my_function_with_args(name, greeting)
```
